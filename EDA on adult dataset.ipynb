**7. Display age statistics for each race (*race* feature) and each gender (*sex* feature). Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.**

```python
# You code here
x = data.groupby(['race','sex']).describe()['age']
x
```

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
# You code here
data['married'] = data['marital-status'].apply(lambda x:1 if x.split('-')[0]=='Married' else 0)
data.head()
```

```python
count = data.groupby(['salary','sex'])['married'].value_counts()
count
```

```python
percent = count.groupby(level=[0,1]).apply(lambda x: x/(x.sum()))
percent
```

**So the answer to the above question is more married men(89.5%) are earning >50k compared to single men(10.4%)**

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
# You code here
print('Maximum no. of hours a person works per week: ',data['hours-per-week'].max())
```

```python
print('Number of people who work for 99 hours per weeek: ',len(data[data['hours-per-week']==99]))
```

```python
count_sal = data[data['hours-per-week']==99].groupby('salary').count()['hours-per-week']
count_sal
```

```python
count_sal/count_sal.sum()
```

**From the above solution we can infer that 29.4% of people who works for 99 hours per week get >50k salary **
